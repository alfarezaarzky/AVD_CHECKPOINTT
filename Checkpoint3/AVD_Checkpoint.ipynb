### **Outliers Values**

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dengan persentase 0,65%

* Tutoring_Sessions

Dengan persentase 6,5%

* Exam_Score

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan kolom diatas, ada beberapa kolom yang memiliki nilai outlier. Yaitu:

* Kolom Hours_Studied dengan persentase 0.65%
* Kolom Tutoring_Sessions dengan persentase 6.50%
* Kolom Exam_Score dengan persentase 1.57%

Maka ini perlu ditangani.

```python
columns_to_impute = ["Hours_Studied", "Tutoring_Sessions", "Exam_Score"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

## **Data Construction**



Data Construction adalah proses membuat atau menambahkan variabel (fitur) baru dari data yang sudah ada untuk memperkaya informasi dan mempermudah analisis.

```python
def categorize_Exam_Score(Exam_Score):
    if Exam_Score > 64:
        return 'A'
    elif Exam_Score < 65:
        return 'B'
    else:
        return 'Break Even'

df['Category_Exam_Score'] = df['Exam_Score'].apply(categorize_Exam_Score)
```

Disini saya menambahkan kolom baru yaitu Category_Exam_Score untuk mendefinisikan nilai ujian siswa (Exam_Score) apakah nilai tersebut termasuk A atau B.
